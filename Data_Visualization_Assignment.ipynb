```python
#4 Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
    'Value': [10, 15, 12, 25, 22, 27, 35, 31, 33, 45, 47, 42]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='coolwarm')

plt.title('Distribution of Numerical Values Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

```python
#5  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

np.random.seed(42)
size = 100
X1 = np.random.randn(size)
X2 = X1 + np.random.randn(size) * 0.5
X3 = 2 * X1 - 1 + np.random.randn(size) * 0.3


data = pd.DataFrame({'Feature_1': X1, 'Feature_2': X2, 'Feature_3': X3})

correlation_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
fig = px.violin(df, x='Grade', y='Score', box=True, points='all')

fig.show()
```

```python
#3Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

fig = px.imshow(heatmap_data, labels={'color':'Average Sales'}, title='Heatmap of Sales Across Days and Months')

fig.show()
```

```python
#4  Using the given x and y data, generate a 3D surface plot to visualize the function
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig = go.Figure(data=[go.Surface(x=df['X'], y=df['Y'], z=df['Z'])])

fig.show()
```

Bokeh Visualization

```python
#1 Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
output_file("sine_wave.html")
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

```python
#2 .Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.models import ColumnDataSource
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 100  # 100 random x-values between 0 and 100
y = np.random.rand(100) * 100  # 100 random y-values between 0 and 100
sizes = np.random.rand(100) * 50 + 10  # Random sizes for markers between 10 and 60
colors = np.random.rand(100) * 255  # Random values for colors between 0 and 255


output_file("scatter_plot.html")

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')

p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, source=source, marker="circle")

show(p)
```

```python
#3  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
#4 Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_file("histogram.html")
p = figure(title="Histogram of Normally Distributed Data",
           x_axis_label='Value',
           y_axis_label='Frequency')

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p)
```

```python
#5  Create a Bokeh heatmap using the provided dataset
from bokeh.palettes import Viridis256  # Import from bokeh.palettes
from bokeh.transform import linear_cmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
output_file("heatmap.html")

p = figure(title="Heatmap",
           x_axis_label='X',
           y_axis_label='Y',
           tools="hover",
           tooltips=[("x", "$x"), ("y", "$y"), ("value", "@image")])

mapper = linear_cmap(field_name="image", palette=Viridis256,
                     low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, palette=Viridis256)

# Show the results
show(p)
```
